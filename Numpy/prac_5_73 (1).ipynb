```python
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
df = pd.read_csv("practical_5.csv", header='infer').values
print(df.shape)
x=df[:,0:-1]
y=df[:,-1]
nrows=df.shape[0]
print(nrows)
trainsplit = float(input("ENTER THE VALUE BETWEEN 0 AND 1: "))
test_size = int((1 - trainsplit) * nrows)
xtrain, xtest, ytrain, ytest = train_test_split(x, y, 
test_size=test_size, shuffle=False)
print("Size of xtrain:", xtrain.shape)
print("Size of ytrain:", ytrain.shape)
print("Size of xtest:", xtest.shape)
print("Size of ytest:", ytest.shape)
```
