```python
# IRIS classification using Distance Weighted KNN Classifier
# Without using sklearn for classification
# Using Euclidean distance as the distance measure

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split


data = pd.read_csv("Iris.csv", header=0).values 


X = data[:, 1:-1]
y = data[:, -1] 


unique_classes = np.unique(y)
class_map = {c: i for i, c in enumerate(unique_classes)}
y_numeric = np.array([class_map[label] for label in y]) 


X_train, X_test, y_train, y_test = train_test_split(X, y_numeric, test_size=0.2, stratify=y_numeric, random_state=42)

n_classes = len(unique_classes)

# Get user input for k
k = int(input("Enter the number of nearest neighbors to be used (k): "))

# Initialize predictions
pred = np.zeros(X_test.shape[0])

# Perform KNN classification
for i in range(X_test.shape[0]):
    # Compute Euclidean distances from test sample to all training samples
    dist = np.sqrt(np.sum((X_train - X_test[i]) ** 2, axis=1))
    
    # Get indices of k nearest neighbors
    k_min_indices = np.argpartition(dist, k)[:k]
    
    # Compute inverse distance weights
    inv_dist = 1 / (dist + 1e-10)  # Avoid division by zero
    denom = np.sum(inv_dist[k_min_indices])

    # Reset class vote counts
    class_votes = np.zeros(n_classes)

    # Weight the votes by inverse distance
    for j in range(k):
        class_votes[y_train[k_min_indices[j]]] += inv_dist[k_min_indices[j]]

    class_votes /= denom  # Normalize votes
    pred[i] = np.argmax(class_votes)  # Assign class with highest weighted vote

# Convert predictions back to original class labels
pred_labels = np.array([unique_classes[int(p)] for p in pred])

# Define accuracy calculation function
def calc_acc(y_pred, y_true):
    return np.sum(y_pred == y_true) / y_pred.shape[0]

# Define precision calculation function (macro-averaged)
def calc_prec(y_pred, y_true):
    classes = np.unique(y_true)
    n_classes = len(classes)
    precision_sum = 0

    for c in classes:
        tp = np.sum((y_pred == c) & (y_true == c))
        fp = np.sum((y_pred == c) & (y_true != c))
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        precision_sum += precision

    return precision_sum / n_classes 
# Evaluate model
accuracy = calc_acc(pred_labels, y_test)
precision = calc_prec(pred_labels, y_test)

print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Precision: {precision * 100:.2f}%")
```
