```python
#Statistical measures using library
import numpy as np
import scipy as st
import statistics as st
n=int(input('Enter how many numbers do you want to work with'))
mylist=[]
#Mean
for i in range(n):
    num=float(input('Enter number'+str(i+1)+':'))
    mylist.append(num)
myarr=np.array(mylist)
avg=np.mean(myarr)
print('The mean of the data is :',avg)
#Variance
print('Variance',np.var(myarr))
#Standard deviation
print('Standard deviation:',np.std(myarr))
#Median
print('Median:',np.median(myarr))
#Mode
print('Mode:',st.mode(myarr))
#Range
mini=np.min(myarr)
maxi=np.max(myarr)
range1=maxi-mini
print('Range:',range1)
```

```python
#Statistical measures without using library
import numpy as np
n=int(input('Enter the no. of elements do you want to store'))
mylist=[]
for i in range(n):
     num=float(input('Enter number'+str(i+1)+':'))
     mylist.append(num)
myarr=np.array(mylist) 
n1=n
#Mean
x_sum=0
for i in myarr:
    x_sum=i+x_sum
mean=x_sum/n1
print('Mean:',mean)
#Variance
square_sum=0
for i in myarr:
    square_sum+=(i-mean)**2
variance=square_sum/n
print('Variance :',variance)
#Standard deviation
std_dev=np.sqrt(variance)
print('Standard deviation',std_dev)
#Median
for i in range(n-1):
    for j in range(i+1,n):
        if myarr[i]>myarr[j]:
            temp=myarr[i]
            myarr[i]=myarr[j]
            myarr[j]=temp
print(myarr)
if(n%2==0):
   med=(myarr[int((n/2)-1)]+myarr[int(n/2)])/2
else:
   med=myarr[n//2]
print('Median:',med)
#Range
min1=myarr[0]
max1=myarr[-1]
range1=max1-min1
print('Range is :',range1)
#Mode
frequency={}
for i in myarr:
    if i in frequency:
        frequency[i]+=1
    else:
        frequency[i]=1
mode_freq=1
for i in frequency.values():
    if i>mode_freq:
        mode_freq=i
mode=[]
for i in frequency:
    if frequency[i]==mode_freq:
        mode.append(i)
if mode_freq==1:
    print('Mode does not exist for given data')
else:
    print('Mode :',mode)
```
