```python
#OUTPUT/CLASS LABEL OF THE DATASET

print(df["Passed"].unique())
```

```python
#CREATING COPY OF ORIGINAL DATASET TO HANDLE MISSING VALUES
df1=df.copy()
#REMOVED ROWS WHERE "PASSED" HAS NULL VALUES

df1 = df.drop(df[df["Passed"].isna()].index)
print(df1)
print(df1.info())
print(df1.shape)
```

```python
import statistics as st

#HANDLING MISSING VALUES BY FILLING EMPTY COLUMNS WITH MEAN AND MODE

#MEAN FOR COLUMN "STUDY HOURS PER WEEK"
mean_study_hours_per_week0 = np.mean(df.loc[(~df["Study Hours per Week"].isna()) & (df["Passed"]=="Yes"), "Study Hours per Week"].values)
mean_study_hours_per_week1=np.mean(df.loc[(~df["Study Hours per Week"].isna()) & (df["Passed"]=="No"), "Study Hours per Week"].values)

print("MEAN 1 (FOR STUDY HOURS PER WEEK): ",mean_study_hours_per_week0)
print("MEAN 2 (FOR STUDY HOURS PER WEEK): ",mean_study_hours_per_week1)

df1.loc[(df1["Study Hours per Week"].isna()) & (df1["Passed"]=="Yes"), "Study Hours per Week"] = mean_study_hours_per_week0
df1.loc[(df1["Study Hours per Week"].isna()) & (df1["Passed"]=="No"), "Study Hours per Week"] = mean_study_hours_per_week1

#MEAN FOR "ATTENDANCE RATE"

mean_attendance_rate0 = np.mean(df.loc[(~df["Attendance Rate"].isna()) & (df["Passed"]=="Yes"), "Attendance Rate"].values)
mean_attendance_rate1=np.mean(df.loc[(~df["Attendance Rate"].isna()) & (df["Passed"]=="No"), "Attendance Rate"].values)

print()
print("MEAN 1 (FOR ATTENDANCE RATE): ",mean_attendance_rate0)
print("MEAN 2 (FOR ATTENDANCE RATE): ",mean_attendance_rate1)

df1.loc[(df1["Attendance Rate"].isna()) & (df1["Passed"]=="Yes"), "Attendance Rate"] = mean_attendance_rate0
df1.loc[(df1["Attendance Rate"].isna()) & (df1["Passed"]=="No"), "Attendance Rate"] = mean_attendance_rate0

#MEAN FOR "PREVIOUS GRADES"

mean_previous_grades0 = np.mean(df.loc[(~df["Previous Grades"].isna()) & (df["Passed"]=="Yes"), "Previous Grades"].values)
mean_previous_grades1=np.mean(df.loc[(~df["Previous Grades"].isna()) & (df["Passed"]=="No"), "Previous Grades"].values)

print()
print("MEAN 1 (FOR PREVIOUS GRADES): ",mean_previous_grades0)
print("MEAN 2 (FOR PREVIOUS GRADES): ",mean_previous_grades1)

df1.loc[(df1["Previous Grades"].isna()) & (df1["Passed"]=="Yes"), "Previous Grades"] = mean_previous_grades0
df1.loc[(df1["Previous Grades"].isna()) & (df1["Passed"]=="No"), "Previous Grades"] = mean_previous_grades1

#MODE FOR "PARTICIPATION IN EXTRACURRICULAR ACTIVITIES"

mode_extracurricular_activities0=df.loc[df["Passed"]=="Yes","Participation in Extracurricular Activities"].mode()[0]
mode_extracurricular_activities1=df.loc[df["Passed"]=="No","Participation in Extracurricular Activities"].mode()[0]

print()
print("MODE 1 (FOR PARTICIPATION IN EXTRACURRICULAR ACTIVITIES) : ",mode_extracurricular_activities0)
print("MODE 2 (FOR PARTICIPATION IN EXTRACURRICULAR ACTIVITIES) : ",mode_extracurricular_activities1)

df1.loc[(df1["Participation in Extracurricular Activities"].isna()) & (df1["Passed"]=="Yes"),"Participation in Extracurricular Activities"]=mode_extracurricular_activities0
df1.loc[(df1["Participation in Extracurricular Activities"].isna()) & (df1["Passed"]=="No"),"Participation in Extracurricular Activities"]=mode_extracurricular_activities1

#MODE FOR "PARENT EDUCATION LEVEL"

mode_parent_education_level0=df.loc[df["Passed"]=="Yes","Parent Education Level"].mode()[0]
mode_parent_education_level1=df.loc[df["Passed"]=="No","Parent Education Level"].mode()[0]

print()
print("MODE 1 (FOR PARENT EDUCATION LEVEL) : ",mode_parent_education_level0)
print("MODE 2 (FOR PARENT EDUCATION LEVEL) : ",mode_parent_education_level1)

df1.loc[(df1["Parent Education Level"].isna()) & (df1["Passed"]=="Yes"),"Parent Education Level"]=mode_parent_education_level0
df1.loc[(df1["Parent Education Level"].isna()) & (df1["Passed"]=="No"),"Parent Education Level"]=mode_parent_education_level1

print()
print("FINAL DATASET AFTER DROPPING COLUMNS HAVING NA VALUES WITH MEAN OR MODE : ")
print()
print(df1.info())
```
