```python
import numpy as np

# XOR truth table
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0, 1, 1, 0])  # Correct XOR outputs

# Number of samples and features
n_samples = X.shape[0]
n_features = X.shape[1]

# Initial weights and bias
w = np.random.uniform(0, 1, size=n_features)
b = np.random.uniform(0, 1)

# Get number of epochs
n_epoch = int(input("Enter the number of epochs: "))

# Learning rate
lr = 0.01

# Training loop
for e in range(n_epoch):  # Epoch loop
    for s in range(n_samples):  # Iterate over each sample
        net = np.dot(X[s, :], w) + b
        a = 1 if net >= 0 else 0
        error = Y[s] - a

        # Update weights and bias using Perceptron Learning Rule
        w += lr * error * X[s, :]
        b += lr * error

# Print final weights and bias
print("Final weights:", w)
print("Final bias:", b)
```
