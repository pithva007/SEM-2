```python
import pandas as pd #for reading iris.csv
import numpy as np
from sklearn.model_selection import train_test_split #for splitting data in train a
from sklearn.metrics import classification_report #we are going to use this for con
#our implementation of precision, recall and f1-measure

data=pd.read_csv("BostonHousing.csv",header='infer').values #reading and converting to numpy

X=data[:,1:-1] #input data
y=data[:,-1] #target data

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y)

nclasses=np.unique(y_train).shape[0]

dist=np.zeros(shape=X_train.shape[0]) #distance vector for each test data, 1 at a t
pred=np.zeros(shape=X_test.shape[0]) #for storing prediction
classvotes=np.zeros(shape=nclasses) #array for storing votes in favour of each class

k=int(input("Enter the number of nearest neighbours to be used, i.e. k:"))

for i in range(X_test.shape[0]): #for loop to process each test case
    dist=np.sqrt(np.sum((X_train-X_test[i])**2,axis=1)) #calculating Euclidean dist
    #between current test record and all training records
    kminind=np.argpartition(dist,k)[0:k] #finding indices of k minimum distances
    invdist=1/(dist+10e-20) #preparing for weight calculation, avoiding divide by 0
    denom=sum(invdist[kminind]) #for weights normalization
    for j in range(k): #calculating votes in favour of each class
    classvotes[int(y_train[kminind[j]])] += invdist[kminind[j]]
    classvotes/=denom #normalizing votes
    #print(classvotes)
    pred[i]=np.argmax(classvotes) #predicted class is the class with the maximum we
#print(pred)
```
