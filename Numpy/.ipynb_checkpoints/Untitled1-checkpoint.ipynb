```python
import pandas as pd
import numpy as np
import math

# Load dataset
data = pd.read_csv("BostonHousing.csv", header='infer').values

x = data[:, 0:-1]
y = data[:, -1]
nrows = data.shape[0]

print("Total Rows:", nrows)

# Get test split from user
test_split = float(input("Enter test split between 0 and 1: "))

# V3

# Compute training set size
nrows_train = math.floor((1 - test_split) * nrows)

# Shuffle indices
all_indices = np.random.permutation(nrows)

# Split data
x_train = x[all_indices[:nrows_train], :]
y_train = y[all_indices[:nrows_train]]

x_test = x[all_indices[nrows_train:], :]
y_test = y[all_indices[nrows_train:]]

# Print shapes
print("Shape:", x_train.shape, y_train.shape, x_test.shape, y_test.shape)

# Ensure proper partitioning
print("Union:", len(set(all_indices[:nrows_train]).union(all_indices[nrows_train:])))
print("Intersection:", len(set(all_indices[:nrows_train]).intersection(all_indices[nrows_train:])))
```
