```python
#Read the Data
import pandas as pd
import time #for measuring time
mat=pd.read_csv("6citytsp.csv", header=None).values

#checking

print(mat)
print(mat.dtype)

from itertools import permutations    #for generating every possible tour

mat=pd.read_csv("6citytsp.csv", header=None).values
citynames=list(range(mat.shape[0]))   #city names, for simplicity 0 to #Cities-1
per=permutations(citynames)           #generating permutations, return type is itertools.perm
print(type(per))                      #checking the type for per
print(list(per))

#print(per) #checking

import numpy as np
st=time.process_time()
besttourlength=np.inf #initially positive infinity
for tour in per:      #examining 1 tour at a time
    tourlength=0
    for i in range(len(tour)-1):
        tourlength+=mat[tour[i], tour[i+1]]
    tourlength+=mat[tour[i+1],tour[0]]
    if tourlength<besttourlength:   #minimum logic
        besttourlength=tourlength
        besttour=tour

et=time.process_time()
time_taken_ms=(et-st)*1000
print("Best Tour: ", besttour)
print("Best Tour's Length: ", besttourlength)
print("Time Taken: ", time_taken_ms)
```

```python
import time
mat=pd.read_csv("6citytsp.csv", header=None).values
#checking
print(mat)
#asking for starting city
startcity=int(input("Enter the start city from 0 to "+ str(mat.shape[0]-1)+ " "))
#generating city names
citynames=list(range(mat.shape[0]))
#removing the starting city
citynames.remove(startcity) #startcity is scanned, so, we need to generate permutations
#generating permutations
per=list(permutations(citynames))
#print(per)

st=time.process_time()
besttourlength=np.inf
for tour in per:
    tourlength=0
    for i in range(len(tour)-1):
        tourlength+=mat[tour[i], tour[i+1]]
tourlength+=mat[tour[i+1], startcity] #lastcity to startcity
tourlength+=mat[startcity, tour[0]] #startcity to actual city
if tourlength < besttourlength: #minimum logic
    besttourlength=tourlength
    besttour=tour

besttour=list(besttour)
besttour.insert(0,startcity)
et=time.process_time()
time_taken_ms=(et-st)*1000
print("Best Tour: ", besttour)
print("Best Tour's Length", besttourlength)
print("Time Taken: ", time_taken_ms)
```
